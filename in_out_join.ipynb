```python

time_worked_year_leave = []
for i in range(4410):
    count = 0
    sum_seconds = 0
    for j in range(249):
        if math.isnan(sub_time[i][j]):
            count = count + 1
        else:
            sum_seconds = sum_seconds + sub_time[i][j]
    time_worked_year_leave.append([i+1, ((sum_seconds/3600)/(249-count)), count])
#     time_worked_year.append([i+1, sum_seconds])
```

```python
with open('time_per_day_and_leaves.csv', 'w') as out_file:
    writer = csv.writer(out_file)
    writer.writerows([["EmployeeID", "Avg_Hours", "No_of_leaves"]])
    for row in range(4410):
        writer.writerows([time_worked_year_leave[row]])
```
