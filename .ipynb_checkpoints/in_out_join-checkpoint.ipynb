# Extracting features from in_time.csv and out_time.csv

```python
import pandas as pd
import numpy as np
import math
import csv
```

### Loading the CSV files 

```python
in_time = pd.read_csv("hr-analytics-case-study/in_time.csv")
out_time = pd.read_csv("hr-analytics-case-study/out_time.csv")
```

### Renaming the unnamed column to "EmployeeID"

```python
in_time.rename(columns = {"Unnamed: 0": "EmployeeID"}, inplace = True) 
out_time.rename(columns = {"Unnamed: 0": "EmployeeID"}, inplace = True) 
```

```python
in_time.head()
```

```python
out_time.head()
```

### Extracting the time from datetime value

```python
for i in range(1,262):
    in_time.iloc[:,i] =  pd.to_datetime(in_time.iloc[:,i]).dt.time
    out_time.iloc[:,i] = pd.to_datetime(out_time.iloc[:,i]).dt.time
```

```python
in_time.head()
```

```python
out_time.head()
```

### Dropping the columns that contain NULL values for all the rows as these are Company Holidays.

```python
in_time = in_time.dropna(axis=1, how="all")
out_time = out_time.dropna(axis=1, how="all")
```

### The in_time and corresponding out_time have NULL values indicating that the employee was on leave on that specific day.

```python
print(in_time.isna().sum()) #250
print(out_time.isna().sum()) #250
```

### 'in_time' data and 'out_time' data stored as numpy arrays for faster computation.

```python
emp_in = in_time.iloc[:,1:]
emp_in = np.array(emp_in)
emp_out = out_time.iloc[:,1:]
emp_out = np.array(emp_out)
```

```python
emp_out.shape
```

### Calculating the total working time of an employee per day (in seconds)

```python
sub_time = []
for i in range(4410):
    t = []
    t.append(i+1)
    for j in range(249):
        diff_h = (emp_out[i][j].hour - emp_in[i][j].hour)*60*60
        diff_m = (emp_out[i][j].minute - emp_in[i][j].minute)
        diff_s = emp_out[i][j].second - emp_in[i][j].second
        time_worked = diff_h + diff_m + diff_s
        t.append(time_worked)
    sub_time.append(t)
```

```python
in_time.columns
```

### The calculated total work time is written to a CSV file.

```python
with open('total_work_time.csv', 'w') as output_file:
    writer = csv.writer(output_file)
    writer.writerows([in_time.columns])
    for row in range(4410):
        writer.writerows([sub_time[row]])
```

### Calculating the average work time and the number of leaves taken.

```python
time_worked_year_leave = []
for i in range(4410):
    count = 0
    sum_seconds = 0
    for j in range(249):
        if math.isnan(sub_time[i][j]):
            count = count + 1
        else:
            sum_seconds = sum_seconds + sub_time[i][j]
    time_worked_year_leave.append([i+1, ((sum_seconds/3600)/(249-count)), count])
```

```python
time_worked_year_leave
```

### Extracted features written to a CSV file.

```python
with open('extracted_features_from_in_out_time.csv', 'w') as out_file:
    writer = csv.writer(out_file)
    writer.writerows([["EmployeeID", "Avg_Hours", "No_of_leaves"]])
    for row in range(4410):
        writer.writerows([time_worked_year_leave[row]])
```
